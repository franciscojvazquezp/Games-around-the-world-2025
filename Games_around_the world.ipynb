Con la finalidad de evitar errores tipográficos en nuestro analísis, vamos a realizar las siguientes acciones en los nombres de 
las columnas:<br>- Reemplazar mayúsculas por minúsculas.<br>- Eliminar espacios al inicio o final.<br>- Reemplazar espacios por 
guión bajo "_".

```python
## Mediante un bucle limpiamos los nombres de las columnas y las mostramos una vez realizada la acción
new_col_names = []
for old_name in df.columns:
    name_stripped= old_name.strip()
    name_lowered= name_stripped.lower()
    name_no_spaces= name_lowered.replace(' ','_')
    new_col_names.append(name_no_spaces)

df.columns= new_col_names

print(df.columns)
```

Ahora conoceremos el tipo de datos que contiene cada columna, así como la cantidad de valores "no nulos" para cada una de ellas.
Esto nos llevará a realizar cambios en el tipo de datos para que esto nos facilite obtener información más adelante.

```python
## Mostramos la información general del DataFrame para conocer datos relevantes
df.info()
```

```python
## Convertimos los datos al tipo más conveniente para el análisis
df['user_score'] = pd.to_numeric(df['user_score'], errors= 'coerce')
df['year_of_release'] = pd.to_numeric(df['year_of_release'], errors='coerce')
df['year_of_release'] = df['year_of_release'].astype('Int64')
df.info()
```

Los datos contenidos en la columna "year_of_release" se han cambiado de float64 a Int64 con la finalidad de mostrar información
más limpia y con ello poder tener un mejor manejo en la obtención de conclusiones más adelante. Adicional a ello, la columna
"user_score" se cambió de tipo 'object' a tipo numérico.

```python
## Mostramos la cantidad de valores faltantes para cada columna para analizar qué acciones realizar al respecto.
print(df.isna().sum())
```

```python
## Generamos una lista con las dos plataformas que analizaremos en la hipótesis.
platforms_to_test = ['PC', 'XOne'] 
## Generamos un DataFrame con la información para las plataformas PC y Xbox One.
df_filtered = df[df['platform'].isin(platforms_to_test)]
## Generamos un elemento Series con los calificaciones de los usuraios para la plataforma PC, además eliminamos valores nulos.
pc_scores = df[df['platform'] == 'PC']['user_score'].dropna()
## Generamos un elemento Series con los calificaciones de los usuraios para la plataforma Xbox One, además eliminamos valores nulos.
xbox_scores = df[df['platform'] == 'XOne']['user_score'].dropna()
## Mostramos la media y la desviación estándar para ambas plataformas.
print(f"PC - Media: {pc_scores.mean():.2f}, Desv. Est: {pc_scores.std():.2f}")
print(f"Xbox - Media: {xbox_scores.mean():.2f}, Desv. Est: {xbox_scores.std():.2f}")
## Asignamos un valor alpha para la realización de la prueba de hipótesis.
alpha= 0.05
t_stat, p_value = st.ttest_ind(pc_scores, xbox_scores, equal_var=True)
print(f"Estadístico t: {t_stat:.4f}")
print(f"Valor p: {p_value:.8f}")

if p_value < alpha:
    print("Rechazamos H₀: Las medias son significativamente diferentes")
else:
    print("No rechazamos H₀: No hay evidencia suficiente de diferencia entre las medias")
```


**Hipótesis nula:** La media de las calificaciones de usuario para las plataformas Xbox One y PC son iguales. <br> Después de realizar los cálculos
correspondientes, se ha comprobado que la **hipótesis nula se rechaza**, ya que el valor p obtenido es menor al valor alpha.
